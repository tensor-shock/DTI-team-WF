```python
def pad_image(image,color,pad_size):
    top=pad_size
    bottom=pad_size
    left=pad_size
    right=pad_size
    padding_color= (color, color, color)
    image = cv2.copyMakeBorder(image, top, bottom, left, right, cv2.BORDER_CONSTANT, value=padding_color)
    return image
```

```python
def resize_down(image):
    down_width = 50
    down_height = 25
    down_points = (down_width, down_height)
    resized_down = cv2.resize(image, down_points, interpolation= cv2.INTER_LINEAR)
    return resized_down
```

```python
def add_label(image,shape_char,x,y,w,h):
    cv2.putText(image,shape_char,(x,y+h), font,
                fontScale, color, thickness, cv2.LINE_AA)
```

```python
cnt=0
for contour in contours:
    # area = cv2.contourArea(contour)
    x,y,w,h = cv2.boundingRect(contour)

    if w*h<area_cutoff :
        continue
    cnt+=1

    subimage=img1[y:y+h,x:x+w]
    subimage=pad_image(subimage,0,10)
    # cv2_imshow(subimage)

    epsilon = threshold_multiplier * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)
    num_vertices = len(approx)
    x-=pad_size
    y-=pad_size

    if num_vertices == 3:
        shape = "Triangle"
        add_label(og_img,'T',x,y,w,h)
    elif num_vertices == 4:
        shape = "Rectangle"
        add_label(og_img,'R',x,y,w,h)
    elif num_vertices == 5:
        shape = "Pentagon"
        add_label(og_img,'P',x,y,w,h)
    elif num_vertices == 6:
        shape = "Hexagon"
        add_label(og_img,'H',x,y,w,h)
    else :
        shape = "Circle"
        add_label(og_img,'C',x,y,w,h)


    shape_coords = (x, y, x + w, y + h)
    cv2.rectangle(og_img,(x,y), (x+w,y+h),(0,255,70),1)
    print(f"Shape: {shape} ,Coordinates: {shape_coords}")

    # cv2_imshow(subimage)



print('Total count', cnt)
```
